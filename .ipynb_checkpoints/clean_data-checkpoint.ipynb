```python
import pandas as pd
import numpy as np
```

```python
data1 = pd.read_csv("data/average_monthly_temperature_by_state_1950-2022.csv")
data2 = pd.read_csv("data/epest_county_estimates.csv")
data3 = pd.read_csv("data/save_the_bees.csv")
data4 = pd.read_csv("data/pollution_2000_2021.csv")
```

```python
group1 = data1.groupby(["year", "state"])
data1["average_temp"] = group1["average_temp"].transform('mean')
data1 = data1[["year", "state", "average_temp", "centroid_lon", "centroid_lat"]]
data1 = data1.drop_duplicates().reset_index()
data1
```

```python
group2 = data2.groupby(["YEAR", "COMPOUND", "STATE_FIPS_CODE"])
data2["EPEST_LOW_KG"] = group2["EPEST_LOW_KG"].transform('mean')
data2["EPEST_HIGH_KG"] = group2["EPEST_HIGH_KG"].transform('mean')
data2 = data2[["COMPOUND", "YEAR", "STATE_FIPS_CODE", "EPEST_LOW_KG", "EPEST_HIGH_KG"]]
data2 = data2[(data2["COMPOUND"] == "2,4-D") |
              (data2["COMPOUND"] == "DICAMBA") |
              (data2["COMPOUND"] == "METRIBUZIN") |
              (data2["COMPOUND"] == "GLYPHOSATE") |
              (data2["COMPOUND"] == "CHLORPYRIFOS")]

data2 = data2.drop_duplicates().reset_index()
data2
```

```python
group3 = data3.groupby(["state", "year"])

data3["num_colonies"] = group3["num_colonies"].transform('mean')
data3["max_colonies"] = group3["max_colonies"].transform('mean')
data3["lost_colonies"] = group3["lost_colonies"].transform('mean')
data3["percent_lost"] = group3["percent_lost"].transform('mean')
data3["added_colonies"] = group3["added_colonies"].transform('mean')
data3["renovated_colonies"] = group3["renovated_colonies"].transform('mean')
data3["percent_renovated"] = group3["percent_renovated"].transform('mean')
data3["other_pests_and_parasites"] = group3["other_pests_and_parasites"].transform('mean')
data3["diseases"] = group3["diseases"].transform('mean')

data3 = data3[["state", 
               "year", 
               "num_colonies", 
               "max_colonies", 
               "lost_colonies", 
               "percent_lost", 
               "added_colonies", 
               "renovated_colonies", 
               "percent_renovated", 
               "other_pests_and_parasites", 
               "diseases"]]

data3 = data3.drop_duplicates().reset_index()
data3
```

```python
group4 = data4.groupby(["Year", "State"]) 

data4["O3 Mean"] = group4["O3 Mean"].transform("mean")
data4["O3 1st Max Value"] = group4["O3 1st Max Value"].transform("mean")
data4["O3 AQI"] = group4["O3 AQI"].transform("mean")
data4["CO Mean"] = group4["CO Mean"].transform("mean")
data4["CO 1st Max Value"] = group4["CO 1st Max Value"].transform("mean")
data4["CO 1st Max Hour"] = group4["CO 1st Max Hour"].transform("mean")
data4["CO AQI"] = group4["CO AQI"].transform("mean")
data4["SO2 Mean"] = group4["SO2 Mean"].transform("mean")
data4["SO2 1st Max Value"] = group4["SO2 1st Max Value"].transform("mean")
data4["SO2 1st Max Hour"] = group4["SO2 1st Max Hour"].transform("mean")
data4["SO2 AQI"] = group4["SO2 AQI"].transform("mean")
data4["NO2 Mean"] = group4["NO2 Mean"].transform("mean")
data4["NO2 1st Max Value"] = group4["NO2 1st Max Value"].transform("mean")
data4["NO2 1st Max Hour"] = group4["NO2 1st Max Hour"].transform("mean")
data4["NO2 AQI"] = group4["NO2 AQI"].transform("mean")

data4 = data4[["Year",
              "State",
              "O3 Mean",
              "O3 1st Max Value",
              "O3 AQI",
              "CO Mean",
              "CO 1st Max Value",
              "CO 1st Max Hour",
              "CO AQI",
              "SO2 Mean",
              "SO2 1st Max Value",
              "SO2 1st Max Hour",
              "SO2 AQI", 
              "NO2 Mean", 
              "NO2 1st Max Value",
              "NO2 1st Max Hour",
              "NO2 AQI"]]

data4 = data4.drop_duplicates().reset_index()
data4
```

```python
data1.to_csv("data/average_monthly_temperature_by_state_1950-2022.csv")
data2.to_csv("data/epest_county_estimates.csv")
data3.to_csv("data/save_the_bees.csv")
```
